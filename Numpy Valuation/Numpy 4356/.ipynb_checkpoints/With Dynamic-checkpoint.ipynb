```python
from PIL import Image
```

```python
import numpy as np
```

```python
im = np.array(Image.open('Flower.jpg'))
im_shape = im.shape
print("Before trimming:",im.shape)
```

```python
a = int((im_shape[0]/2)-250)
b = int((im_shape[0]/2)+250)
c = int((im_shape[1]/2)-250)
d = int((im_shape[1]/2)+250)

im_trim = im[a:b, c:d]
print("After trimming:",im_trim.shape)
```

```python
Image.fromarray(im_trim).save('trim_Flower.jpg')
```

```python
from matplotlib import image
from matplotlib import pyplot
```

```python
t_image = image.imread('trim_Flower.jpg')
```
